### 교육

```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import time
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook

url = "http://bokjiro.go.kr/welInfo/retrieveWelInfoBoxList.do"

driver = wb.Chrome()
driver.get(url)

title_list = []
content_list = []
support_list = []

driver.find_element_by_css_selector("li.cate11").click()
driver.find_element_by_css_selector("select#pageUnit>option:nth-child(3)").click()
driver.find_element_by_css_selector("a.btnInput").click()
for num in range(1,73):
    
    # 이게 a링크 타고 들어가는거
    a = "ul.full.type03.listStyle>li:nth-child("+str(num)+")>div.cardStyleIn>dl>dt.tit>a"
    menu = driver.find_element_by_css_selector(a)
    
    menu.click()
    time.sleep(1)
   
    # title 빼오기
    try:
        title = driver.find_element_by_css_selector('span.type01').text
    except:
        title = "제목을 찾을 수 없습니다."
    
    
    # content 빼오기
    try:
        content = driver.find_element_by_css_selector('div.cont').text
    except:
        content = "내용을 찾을 수 없습니다."
        
      # support 빼오기
    try:
        support = driver.find_element_by_css_selector('ul.bokjiContsIn>li.first').text
    except:
        support = "지원대상을 찾을 수 없습니다."
        
    
    title_list.append(title)
    content_list.append(content)
    support_list.append(support)
        
    
    
    driver.back()
    time.sleep(2)
    
driver.close()
print(title_list)
print(content_list)
print(support_list)


```
