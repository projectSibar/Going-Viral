```python
for i in fileList : 
    if i.select(' a'):
        beforelink = i.select(' a')
        afterList.append(beforelink[0]['href'])
    else:
        afterList.append(i.text.strip())

dic = {'title':titleList,'date':dateList,'fileLink':afterList,'contents':contentList}
crawing3=pd.DataFrame(dic)
#csv파일로 저장하기
nowDateTime = now.strftime('%Y-%m-%d')
crawing3.to_csv('광주서구통합도서관'+nowDateTime+'.csv',encoding = "utf-8")
```
