# 광주 보건소

```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import time
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook


title_list = []
content_list = []


for x in range(1,75):
    url = "https://health.gwangsan.go.kr/bb/bbBoard.php?movePage="+str(x)+"&action=list&pageID=health0706000000&boardID=BNEWS"

    driver = wb.Chrome()
    driver.get(url)


    for num in range(1,10):

        # 이게 a링크 타고 들어가는거
        a = "tbody tr:nth-child("+str(num)+") a"
        menu = driver.find_element_by_css_selector(a)

        menu.click()
        time.sleep(1.5)

        # title 빼오기
        try:
            title = driver.find_element_by_css_selector('div.board_view>h4').text
        except:
            title = "제목을 찾을 수 없습니다."

        # content 빼오기
        try:
            content = driver.find_element_by_css_selector('div.board_viewDetail').text
        except:
            content = "내용을 찾을 수 없습니다."




        title_list.append(title)
        content_list.append(content)


    
        driver.back()
        time.sleep(1.5)
    print(x)
    driver.close()
print(title_list)
print(content_list)
```

```python
title_list_ver1 = []

for i in title_list:
    temp = i.replace("\n"," ")
    title_list_ver1.append(temp)

print(title_list_ver1)
```

```python
dic = {'title': title_list_ver1, 'content': content_list_ver1}

df = pd.DataFrame(dic,columns=['title','content'])
df.to_csv('3rd_project_health001.csv', index = False, encoding='UTF-8')
```
